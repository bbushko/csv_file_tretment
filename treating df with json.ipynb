```python
import pandas as pd
import json
import re
```

```python
df = pd.read_csv('%FULL PATH TO FILE%')

def parse_json(df):
    """
    Firstly, replacing the list of dictionaries with dictionary of dictionaries, because there are duplicate columns
    Then replacing all malicios pattern and rewriting them..
    After it replacing string representation of the dictionary with actual dictionary
    Finally, concatenating our df with normalized json.
    """
    
    #  These regex patterns are used for making dictionary appropriate for json parsing
    pattern_account = r'<AccountType\..+?: ' 
    pattern_decimal = r'Decimal\('

    count = 0
    for elem in df['journal_lines']:
        res = elem.replace("[{", '{"1": {').replace('}]', '}}').replace('}, {', '}, "2": {')  # a lot of patterns here
        res = re.sub(pattern_account, '', string=res).replace('>', '').replace("'", '"').replace('None', '""')
        res = re.sub(pattern_decimal, '', string=res).replace('")', '"').replace('s" ', "s' ")
        res = json.loads(res) 
        df.iat[count,9] = res
        count += 1  # descending by rows of a dataframe and assigning it the new json value

    df = pd.concat([df, pd.json_normalize(df['journal_lines'])], axis=1)  # flattening the json file
    df.to_csv('journal_data.csv')
    return df

parse_json(df)
```
